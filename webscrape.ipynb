```python
# importing libraries
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
import os
```

```python
# get the response in the form of html
wikiurl="https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table"
table_class="wikitable sortable jquery-tablesorter" # to get table from a website
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
indiatable=soup.find('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(indiatable))
# convert list to dataframe
df=pd.DataFrame(df[0])
df
```

```python
medals_df = df.copy()
medals_df
```

```python
medals_df = medals_df.rename(columns={
                            ".mw-parser-output .tooltip-dotted{border-bottom:1px dotted;cursor:help}Team (IOC code)": "Country"
                                })
```

```python
medals_df
```

```python
medals_df.to_csv("Resources/medals_df_temp.csv",header=None,index=False)
```

```python
col_Names=["country","summer_participations","summer_gold","summer_silver","summer_bronze","summer_total","winter_participations","winter_gold","winter_silver","winter_bronze","winter_total","total_participation","total_gold","total_silver","total_bronze","total_total "]
medals_df = pd.read_csv("Resources/medals_df_temp.csv", names=col_Names)
os.remove("Resources/medals_df_temp.csv")
medals_df
```

```python
medal_totals_df = medals_df[['country', 'summer_total', 'winter_total', 'total_participation', 'total_total ']].copy()
medal_totals_df = medal_totals_df.rename(columns={'total_total ': 'total_won'})
medal_totals_df['country'] = medal_totals_df['country'].map(lambda country: country[:country.find('(')-1] if country.find('(') > -1 else country)
medal_totals_df.head()
```

```python
medal_totals_df.to_csv("Resources/olympics_medals_country_wise.csv",index=False)
```
